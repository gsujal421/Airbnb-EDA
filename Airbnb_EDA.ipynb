#Exploratory Data Analysis

```python
# Identifying Outlier
sns.boxplot(data=df,y='price')
```

```python
# Outlier Removal using IQR Method

Q1 = df['price'].quantile(0.25)
Q3 = df['price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df_no_outliers = df[(df['price'] >= lower_bound) & (df['price'] <= upper_bound)]

print(f"Original dataset shape: {df.shape}")
print(f"After removing outliers: {df_no_outliers.shape}")

plt.figure(figsize=(6, 4))
sns.boxplot(data=df_no_outliers, y='price')
plt.title('Price Distribution after Removing Outliers (IQR Method)')
plt.show()
```

```python
#Trend Analysis: Reviews Over Time

# Ensure last_review is datetime
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')

monthly_reviews = df.set_index('last_review').resample('M')['number_of_reviews'].sum().reset_index()

# Plot trend
plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_reviews, x='last_review', y='number_of_reviews', marker='*', color='red')
plt.title('Trend of Reviews Over Time')
plt.xlabel('Month')
plt.ylabel('Total Number of Reviews')
plt.tight_layout()
plt.show()
```
